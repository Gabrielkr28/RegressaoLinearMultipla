```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Função para calcular o coeficiente de correlação
def correlacao(x, y):
    r = np.corrcoef(x, y)[0, 1]
    return r

# Função para calcular a regressão linear (β0 e β1)
def regressao(x, y):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    return intercept, slope

# Função para criar o gráfico de dispersão
def criar_grafico_dispersao(x, y, dataset_nome):
    plt.scatter(x, y, label=dataset_nome)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)

# Função para mostrar os coeficientes de correlação e regressão no gráfico
def mostrar_coeficientes(r, beta0, beta1):
    plt.title(f'Coeficiente de Correlação: {r:.2f}\nRegressão: y = {beta1:.2f}x + {beta0:.2f}')

# Datasets de exemplo
dataset1_x = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
dataset1_y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])

dataset2_x = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
dataset2_y = np.array([9.14, 8.14, 8.47, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])

dataset3_x = np.array([8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 19])
dataset3_y = np.array([6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 5.56, 7.91, 6.89, 12.50])

# Plote e analise cada dataset
plt.figure(figsize=(12, 5))

plt.subplot(2, 2, 1)
criar_grafico_dispersao(dataset1_x, dataset1_y, 'Dataset 1')
r = correlacao(dataset1_x, dataset1_y)
beta0, beta1 = regressao(dataset1_x, dataset1_y)
mostrar_coeficientes(r, beta0, beta1)
plt.plot(dataset1_x, beta0 + beta1 * dataset1_x, color='red')

plt.subplot(2, 2, 2)
criar_grafico_dispersao(dataset2_x, dataset2_y, 'Dataset 2')
r = correlacao(dataset2_x, dataset2_y)
beta0, beta1 = regressao(dataset2_x, dataset2_y)
mostrar_coeficientes(r, beta0, beta1)
plt.plot(dataset2_x, beta0 + beta1 * dataset2_x, color='red')

plt.subplot(2, 2, 3)
criar_grafico_dispersao(dataset3_x, dataset3_y, 'Dataset 3')
r = correlacao(dataset3_x, dataset3_y)
beta0, beta1 = regressao(dataset3_x, dataset3_y)
mostrar_coeficientes(r, beta0, beta1)
plt.plot(dataset3_x, beta0 + beta1 * dataset3_x, color='red')

plt.tight_layout()
plt.show()
```
